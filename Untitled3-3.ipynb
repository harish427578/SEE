```python
import matplotlib.pyplot as plt
import numpy as np

try:
    with open("coefficients.txt", "r") as file:
        a, b, c = [float(line.strip()) for line in file.readlines()]
except FileNotFoundError:
    print("coefficients.txt not found. Using default values.")
    a, b, c = 1, -3, 2

x = np.linspace(-10, 10, 100)
y = a * x**2 + b * x + c

plt.plot(x, y, label=f"y = {a}x² + {b}x + {c}")
plt.title("Weather Modeling - File Input")
plt.xlabel("Time (x)")
plt.ylabel("Weather Metric (y)")
plt.legend()
plt.grid()
plt.show()
```
