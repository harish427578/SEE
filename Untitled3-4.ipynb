```python
import matplotlib.pyplot as plt
import numpy as np

try:
    with open("multi_coefficients.txt", "r") as file:
        datasets = [list(map(float, line.strip().split())) for line in file]
except FileNotFoundError:
    print("multi_coefficients.txt not found. Using default values.")
    datasets = [[1, -3, 2]]

x = np.linspace(-10, 10, 100)

plt.figure(figsize=(10, 6))
for idx, (a, b, c) in enumerate(datasets, start=1):
    y = a * x**2 + b * x + c
    plt.plot(x, y, label=f"Set {idx}: y = {a}x + {b}x + {c}")

plt.title("Weather Modeling - Multiple Sets of Inputs")
plt.xlabel("Time (x)")
plt.ylabel("Weather Metric (y)")
plt.legend()
plt.grid()
plt.show()
```
