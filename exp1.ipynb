```python
import matplotlib.pyplot as plt
import numpy as np
a, b, c = 1, -3, 2

x = np.linspace(-10, 10, 100)
y = a * x**2 + b * x + c

plt.plot(x, y, label=f"y = {a}x² + {b}x + {c}")
plt.title("Weather Modeling - Hardcoded Variables")
plt.xlabel("Time (x)")
plt.ylabel("Weather Metric (y)")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    with open("multi_coefficients.txt", "r") as file:
        datasets = [list(map(float, line.strip().split())) for line in file]
except FileNotFoundError:
    print("multi_coefficients.txt not found. Using default values.")
    datasets = [[1, -3, 2]]

x = np.linspace(-10, 10, 100)

plt.figure(figsize=(10, 6))
for idx, (a, b, c) in enumerate(datasets, start=1):
    y = a * x**2 + b * x + c
    plt.plot(x, y, label=f"Set {idx}: y = {a}x² + {b}x + {c}")

plt.title("Weather Modeling - Multiple Sets of Inputs")
plt.xlabel("Time (x)")
plt.ylabel("Weather Metric (y)")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(a, b, c, x_range=(-10, 10)):
    x = np.linspace(x_range[0], x_range[1], 100)
    y = a * x**2 + b * x + c
    return x, y

def plot_model(data_sets, title):
    plt.figure(figsize=(10, 6))
    for idx, (a, b, c) in enumerate(data_sets, start=1):
        x, y = quadratic_model(a, b, c)
        plt.plot(x, y, label=f"Set {idx}: y = {a}x² + {b}x + {c}")
    plt.title(title)
    plt.xlabel("Time (x)")
    plt.ylabel("Weather Metric (y)")
    plt.legend()
    plt.grid()
    plt.show()

data_sets = [(1, -3, 2), (2, -4, 1), (-1, 3, -2)]
plot_model(data_sets, "Weather Modeling - Modular Design")
```
