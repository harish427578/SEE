```python
import matplotlib.pyplot as plt
import numpy as np
a, b, c = 1, -3, 2

x = np.linspace(-10, 10, 100)
y = a * x**2 + b * x + c

plt.plot(x, y, label=f"y = {a}x² + {b}x + {c}")
plt.title("Weather Modeling - Hardcoded Variables")
plt.xlabel("Time (x)")
plt.ylabel("Weather Metric (y)")
plt.legend()
plt.grid()
plt.show()
```
